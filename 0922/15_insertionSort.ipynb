```python
sortData = [8, 3, 4, 9, 1]
for i in range(1, len(sortData)):
    # 정렬된 파일에 새로 새로운 레코드의 위치를 찾기 위해 새로운 레코드와 인덱스를 각각 key와 index에 저장한다.
    key = sortData[i]
    index = i
    # print(key, index)
    
    # 새로운 레코드가 삽입될 위치를 찾는다.
    for j in range(i - 1, -1, -1):
        # print('i = {}, j = {}'.format(i, j))
        if sortData[j] > key:
            # 앞의 데이터가 크기 때문에 삽입할 레코드의 위치로 이동시키고 삽입할 레코드의 인덱스를 1감소 시킨다.
            sortData[j + 1] = sortData[j]
            index -= 1
        else:
            # 삽입할 레코드가 이미 정렬된 데이터보다 크다면 더 이상 앞의 데이터와 비교할 필요가 없으므로 반복을 탈출한다.
            break # sortData[index] = key를 실행하러 간다.
        # ============ if sortData[j] > key: 끝
    # ============ for j in range(i - 1, -1, -1): 끝 => 회전 끝
    
    # 새로운 레코드를 위치에 맞게 삽입한다.
    sortData[index] = key
    print('{}회전 결과 : {}'.format(i, sortData))
# ============ for i in range(1, len(sortData)): 끝 => 정렬 끝
# print('정렬 결과 : {}'.format(sortData))
```
